- NaÃ¯ve Bayes ë¶„ë¥˜ê¸°ëŠ” ì§€ë„í•™ìŠµì´ë¼ì„œ ì •ë‹µì„ ì•Œë ¤ì¤˜ì•¼ í•œë‹¤.

### --â–¶ English

- ë¨¼ì € ì˜ì–´ë¡œ~, ì˜ì–´ëŠ” nltkë§Œ ìˆìœ¼ë©´ ëœë‹¤.

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
train = [
	("i like you", "pos"),
	("i hate you", "neg"),
	('you like me', 'neg'),
	('i like her', 'pos')
]
```

```python
train[0][0]
```

```python
sentence = train[0]
word_tokenize(sentence[0])
```

```python
[word.lower() for sentence in train for word in word_tokenize(sentence[0])]
```

```python
# ì „ì²´ ë§ ë­‰ì¹˜
all_words = set(
	word.lower() for sentence in train for word in word_tokenize(sentence[0])
)
all_words
```

```python
# ë§ ë­‰ì¹˜ ëŒ€ë¹„í•´ì„œ trainì˜ ê° sentenceì— ë‹¨ì–´ê°€ ìˆê³  ì—†ìŒì„ í‘œê¸°í•œë‹¤.

t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train]
t
```

#### â—¾ ë§ ë­‰ì¹˜ í•™ìŠµ

ğŸ‘‰ í•´ì„

- 'hate'ë€ ë‹¨ì–´ê°€ ì—†ìœ¼ë©´(False) ê¸ì •(pos)ì¼ í™•ë¥ ì´ 1.7ëŒ€ 1

- 'like'ê°€ ìˆì„ ë•Œ, Positiveì¼ í™•ë¥ ì´ 1.7ëŒ€ 1ì´ë‹¤.

- ê° ë‹¨ì–´ë³„ë¡œ ë…ë¦½ì ìœ¼ë¡œ í™•ë¥ ì„ ê³„ì‚°í•˜ê¸° ë•Œë¬¸ì— naÃ¯veí•˜ë‹¤ë¼ê³  í•œë‹¤.

```python
clf = nltk.NaiveBayesClassifier.train(t)
clf.show_most_informative_features()
```

#### â—¾ nltk ë¶„ë¥˜ê¸° í…ŒìŠ¤íŠ¸

```python
test_sentence = "i like MeRui"
test_sent_features = {
	word.lower(): (word in word_tokenize(test_sentence.lower())) for word in all_words
}
test_sent_features
```

```python
clf.classify(test_sent_features)
```

### --â–¶ Korean

- ì •ë‹µì„ ì•Œê³  ìˆëŠ”('pos' or 'neg') ë¬¸ì¥ìœ¼ë¡œ í›ˆë ¨ìš© ë°ì´í„°ë¥¼ ì¤€ë‹¤.

```python
from konlpy.tag import Okt
pos_tagger = Okt()
```

```python
train = [
	("ë©”ë¦¬ê°€ ì¢‹ì•„", 'pos'),
	("ê³ ì–‘ì´ë„ ì¢‹ì•„", 'pos'),
	("ë‚œ ìˆ˜ì—…ì´ ìë£¨í•´", 'neg'),
	("ë©”ë¦¬ëŠ” ì´ìœ ê³ ì–‘ì´ì•¼", 'pos'),
	("ë‚œ ë§ˆì¹˜ê³  ë©”ë¦¬ë‘ ë†€ê±°ì•¼", 'pos'),
]
```

#### â—¾ ì „ì²´ ë§ ë­‰ì¹˜ ë§Œë“¤ê¸°

- 'ë©”ë¦¬ê°€', 'ë©”ë¦¬ëŠ”', 'ë©”ë¦¬ë‘'ì„ ëª¨ë‘ ë‹¤ë¥¸ ë‹¨ì–´ë¡œ ì¸ì‹í•˜ëŠ” ë¬¸ì œê°€ ë°œìƒí•œë‹¤.

- í•œê¸€ì€ í˜•íƒœì†Œ ë¶„ì„ì´ í•„ìˆ˜ì´ë‹¤.!!

```python
all_words = set(
	word.lower() for sentence in train for word in word_tokenize(sentence[0])
)
all_words
```

```python
t = [
	({word: (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train
]
t
```

#### â—¾ ë§ ë­‰ì¹˜ í•™ìŠµ

```python
clf = nltk.NaiveBayesClassifier.train(t)
clf.show_most_informative_features()
```

#### â­ í˜•íƒœì†Œ ë¶„ì„ : í•¨ìˆ˜ ìƒì„±

- í•œê¸€ì€ í˜•íƒœì†Œ ë¶„ì„ì„ í•„ìˆ˜ì ìœ¼ë¡œ í•´ì•¼ ì œëŒ€ë¡œ ëœ ê²°ê³¼ë¥¼ ë„ì¶œí•  ìˆ˜ ìˆë‹¤.

- í˜•íƒœì†Œ ë¶„ì„ì„ í•œ í›„ í’ˆì‚¬ë¥¼ ë‹¨ì–´ ë’¤ì— ë¶™ì´ëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜í•œë‹¤.

```python
def tokenize(doc):
    return ["/".join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]
```

```python
train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

```python
tokens = [t for d in train_docs for t in d[0]]
tokens
```

```python
def term_exists(doc):
    return {word: (word in set(doc)) for word in tokens}
```

```python
train_xy = [(term_exists(d), c) for d, c in train_docs]
train_xy
```

#### â—¾ ë§ ë­‰ì¹˜ í•™ìŠµ

```python
clf = nltk.NaiveBayesClassifier.train(train_xy)
clf.show_most_informative_features()
```

#### â—¾ nltk ë¶„ë¥˜ê¸° í…ŒìŠ¤íŠ¸

```python
test_sentence = [("ë‚œ ìˆ˜ì—…ì´ ë§ˆì¹˜ë©´ ë©”ë¦¬ë‘ ë†€ê±°ì•¼")]

test_docs = pos_tagger.pos(test_sentence[0])
test_docs
```

```python

test_sent_features = {word: (word in tokens) for word in test_docs}
test_sent_features
```

```python
clf.classify(test_sent_features)
```
