```python
import nltk
from konlpy.corpus import kobill

# doc_ko = kobill.open("1809890.txt").read()
with kobill.open("1809890.txt") as kb:
    doc_ko = kb.read()

print(doc_ko)
```

#### --▶ Okt 엔진으로 명사 분석

```python
from konlpy.tag import Okt

t = Okt()
tokens_ko = t.nouns(doc_ko)
print(tokens_ko, len(tokens_ko))
```

#### --▶ nltk로 토큰(빈도수 포함) 분석

- token
	- (화폐 대용으로 쓰는) 토큰/ 상품권,교환권/ 선물권,할인권/ 표시,징표/ 집단\[전체]를 대표하는 것, 견본
	- [언어학] 토큰:
		- 단어 한 개 혹은 구두점 한 개를 지칭하는 말뭉치(corpus)의 최소 단위
		- 특정 낱말·표현·문장이 실제로 쓰이는 예/횟수<br></br>

- `nltk.Text`(tokens=`Okt().nouns(doc)`, name='').vocab() --> 개별 토큰의 빈도수를 dict{'token':num,...} 형태로 반환한다.

- `nltk.FreqDist`(token_list) --> 토큰 리스트 내 형태소 각각의 빈도수를 FreqDist({'token':num,...}) 형태로 반환한다.

## 3️⃣ 나이브베이즈 : 감성 분석

> 나이브 베이즈 분류(Naive Bayes Classifier)를 이용한 감성분석

- Naive Bayes Classifier는 특성들 사이의 독립을 가정하는 베이즈 정리를 적용한 확률 분류기의 일종이다.

- Naïve Bayes 분류기는 지도학습이라서 정답을 알려줘야 한다.

### --▶ English

- 먼저 영어로~, 영어는 nltk만 있으면 된다.

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
# Naive Bayes Classifier는 지도 학습으로 정답(긍정적 혹은 부정적)을 달아주어여 한다.

train = [
	("i like you", "pos"),
	("i hate you", "neg"),
	('you like me', 'neg'),
	('i like her', 'pos')
]
```

```python
train[0][0]
```

```python
sentence = train[0]
word_tokenize(sentence[0])
```

```python
# 전체 말 뭉치를 만든다.
# set container를 적용하지 않으면 중복된 모든 말뭉치를 포함한다. 

[word.lower() for sentence in train for word in word_tokenize(sentence[0])]
```

```python
# 전체 말 뭉치
all_words = set(
	word.lower() for sentence in train for word in word_tokenize(sentence[0])
) #--> sentence[1]은 'pos', 'neg' 라벨임.
all_words
```

```python
# 전체 말 뭉치에 train의 각 sentence의 단어가 있고 없음을 표기한다.
# 모든 단어의 존재 유무가 50:50이므로 특성들 사이의 독립을 가정할 수 있다.
# 이렇게 각 단어별로 독립직으로 확률을 계산하기 때문에 naïve하다고 한다.

t = [({word: (word in word_tokenize(sentence[0])) for word in all_words}, x[1]) for sentence in train]
t
```

#### ◾ 말 뭉치 학습

- sklearn에서 모델의 fit()을 nltk에서는 train()으로 모델이 학습한다.

👉 해석

- 'hate'란 단어가 없으면(False) 긍정(pos)일 확률이 1.7대 1

- 'like'가 있을 때, Positive일 확률이 1.7대 1이다.

- 각 단어별로 독립적으로 확률을 계산하기 때문에 naïve하다라고 한다.

```python
clf = nltk.NaiveBayesClassifier.train(t)
clf.show_most_informative_features()
```

#### ◾ nltk 분류기 예측 : 감성분석

- predict()와 동일한 기능의 예측은 classify()로 한다.

```python
test_sentence = "i like MeRui"
test_sent_features = {
	word.lower(): (word in word_tokenize(test_sentence.lower())) for word in all_words
}
test_sent_features
```

```python
clf.classify(test_sent_features)
```

#### ⭐ 형태소 분석 : 함수 생성

- 한글은 형태소 분석을 필수적으로 해야 제대로 된 결과를 도출할 수 있다.

- 형태소 분석을 한 후 품사를 단어 뒤에 붙이는 함수를 정의한다.

```python
def tokenize(doc):
    return ["/".join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]
```

```python
train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

```python
tokens = [t for doc in train_docs for t in doc[0]]
tokens
```

```python
def term_exists(doc):
    return {word: (word in set(doc)) for word in tokens}
```

```python
train_xy = [(term_exists(token), label) for token, label in train_docs]
train_xy
```

#### ◾ 말 뭉치 학습

```python
clf = nltk.NaiveBayesClassifier.train(train_xy)
clf.show_most_informative_features()
```

#### ◾ nltk 분류기 예측 : 감성분석

```python

test_sent_features = {word: (word in tokens) for word in test_docs}
test_sent_features
```

### --▶ __tf-idf vectorize__

- TF-IDF(Term Frequency - Inverse Document Frequency)

- 정보 검색과 텍스트 마이닝에서 사용하는 가중치로 여러 문서로 이루어진 문서군이 있을 때,</br>
	어떤 단어가 특정 문서 내에서 얼마나 중요한 것인지를 나타내는 통계적 수치이다.

- 참고: https://ko.wikipedia.org/wiki/Tf-idf 

- 한 문서에서 많이 등장한 단어에 가중치를 부여하고, --> TF(Term Freq.)

- 전체 문서에서 많이 나타나는 단어는 중요하지 않게 가중치를 부여한다. --> IDF(Inverse Document Freq.) <br></br>

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(min_df=1, decode_error='ignore')
```

```python
X = tfidf_vectorizer.fit_transform(contents_for_vectorize)
X.shape
```

```python
X.toarray().transpose()
```

```python
new_post_vector = tfidf_vectorizer.transform(new_post_for_vectorize)
new_post_vector.toarray()
```

```python
def dist_norm(v1, v2):
    # 벡터의 크기를 모두 1로 정규화 --> 하나의 특성(단어)가 두드러지는 것을 방지할 수 있다.
    v1_normalized = v1 / sp.linalg.norm(v1.toarray())
    v2_normalized = v2 / sp.linalg.norm(v2.toarray())
    
    delta = v1_normalized - v2_normalized
    
    return sp.linalg.norm(delta.toarray())
```

```python
dist = [dist_norm(each, new_post_vector) for each in X]
dist
```

```python
print('Best post is ', dist.index(min(dist)), ', dist =', min(dist))
print('Test post -->', new_post)
print('Best dist post -->', contents[dist.index(min(dist))])
```

#### ◾ `gen_search_url()` / `get_result_onepage()`

```python
import json
import datetime
import urllib.request

def gen_search_url(api_node, search_text, start_num, disp_num):
    """
    - api_node : 쇼핑, 블로그, 카페 등 어디에서 검색할 것인지
    - search_text : 검색어
    - start_num : 검색 시작 위치
    - disp_num : 한 번에 표시할 검색 결과 개수
    """
    base = "https://openapi.naver.com/v1/search"
    node = "/" + api_node + ".json"
    param_query = "?query=" + urllib.parse.quote(search_text)
    param_start = "&start=" + str(start_num)
    param_disp = "&display=" + str(disp_num)
    
    return base + node + param_query + param_start + param_disp


def get_result_onepage(url):
    # client_id = "LGEjycuWmA1fDTIGyyam"
    # client_secret = "3ie0bSi2z_"

    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id", client_id)
    request.add_header("X-Naver-Client-Secret", client_secret)
    
    response = urllib.request.urlopen(request)
    
    print("[%s] URL Request Success" % datetime.datetime.now())
    
    return json.loads(response.read().decode("utf-8"))
```

```python
def tfidf(t, d, D):
    tf = float(d.count(t) / sum(d.count(w)) for w in set(d))
    idf = sp.log(float(len(D)) / (len([doc for doc in D if t in doc])))
    return tf * idf
```
