# Decision TreeÏùò Î∂ÑÌï† Í∏∞Ï§Ä

## üî∞ Î∂ÑÌï† Í∏∞Ï§Ä

> Split Criterion of Decision Tree

### ‚èπ Ï†ïÎ≥¥ ÏóîÌä∏Î°úÌîº

> **Information Entropy**

- ÏóîÌä∏Î°úÌîº Í∞íÏù¥ ÎÇÆÏùÑÏàòÎ°ù ClassÎ≥ÑÎ°ú Î∂ÑÎ•òÍ∞Ä Ïûò Îêú Í≤ÉÏù¥Îã§.

```python
import numpy as np

p = np.arange(0.001, 1, 0.001)
plt.grid()
plt.title('$-p_i\cdot\log_2p_i$')
plt.plot(p, -p*np.log2(p));
```

‚îî‚ñ∂ Î∂ÑÌï† Ï†Ñ

- $~\text{Entropy}=\displaystyle\sum_{i=1}^{n}-p_i\log_2p_i$

	- ÌëúÎ≥∏Í≥µÍ∞Ñ: $n=16$
	- ÌååÎûÄÍ≥µ 6Í∞ú
	- Îπ®Í∞ÑÍ≥µ 10Í∞ú

$~~~~~~~~~~~\Rightarrow~\text{Entropy}=\left(-\frac{~6~}{16}\log_2\frac{~6~}{16}\right)+\left(-\frac{~10~}{16}\log_2\frac{~10~}{16}\right)$

```python
(-(10/16)*np.log2(10/16)) + (-(6/16)*np.log2(6/16))
```

‚îî‚ñ∂ Î∂ÑÌï† ÌõÑ

- $~\text{Entropy}=\displaystyle\sum_{k=1}^dR_k\{\sum_{i=1}^{n}-p_i\log_2p_i\}$

	<img src="https://github.com/ElaYJ/Study_Machine_Learning/assets/153154981/dff8644d-60a8-4baf-ad8e-777165c1b4fd" width="30%" height="30%"></br>

	- Í∑∏Î£π 1. ÌëúÎ≥∏Í≥µÍ∞Ñ $n=8~(R_1=\frac{~8~}{16}),~$ ÌååÎûÄ Í≥µ 1Í∞ú, Îπ®Í∞Ñ Í≥µ 7Í∞ú
	
	- Í∑∏Î£π 2. ÌëúÎ≥∏Í≥µÍ∞Ñ $n=8~(R_2=\frac{~8~}{16}),~$ ÌååÎûÄ Í≥µ 5Í∞ú, Îπ®Í∞Ñ Í≥µ 3Í∞ú

$~~~~~~~~~~~\Rightarrow~\text{Entropy}=0.5\{\left(-\frac{~1~}{8}\log_2\frac{~1~}{8}\right)+\left(-\frac{~7~}{8}\log_2\frac{~7~}{8}\right)\} + 0.5\{\left(-\frac{~5~}{8}\log_2\frac{~5~}{8}\right)+\left(-\frac{~3~}{8}\log_2\frac{~3~}{8}\right)\}$

```python
0.5*(-(1/8)*np.log2(1/8)-(7/8)*np.log2(7/8)) + 0.5*(-(5/8)*np.log2(5/8)-(3/8)*np.log2(3/8))
```

### ‚èπ ÏßÄÎãà Í≥ÑÏàò

> **Gini Index**



‚îî‚ñ∂ Î∂ÑÌï† Ï†Ñ

- $~\text{Gini}=\displaystyle 1-\sum_{i=1}^{n}{p_i}^2$

	- ÌëúÎ≥∏Í≥µÍ∞Ñ: $n=16$
	- ÌååÎûÄÍ≥µ 6Í∞ú
	- Îπ®Í∞ÑÍ≥µ 10Í∞ú

$~~~~~~~~~~~\Rightarrow~\text{Gini}=1-\left((\frac{~6~}{16})^2 + (\frac{~10~}{16})^2\right)$

```python
1 - ((6/16)**2 + (10/16)**2)
```

‚îî‚ñ∂ Î∂ÑÌï† ÌõÑ

- $~\text{Gini}=\displaystyle\sum_{k=1}^dR_k\{1 - \sum_{i=1}^{n}{p_i}^2\}$

	<img src="https://github.com/ElaYJ/Study_Machine_Learning/assets/153154981/dff8644d-60a8-4baf-ad8e-777165c1b4fd" width="30%" height="30%"></br>

	- Í∑∏Î£π 1. ÌëúÎ≥∏Í≥µÍ∞Ñ $n=8~(R_1=\frac{~8~}{16}),~$ ÌååÎûÄ Í≥µ 1Í∞ú, Îπ®Í∞Ñ Í≥µ 7Í∞ú
	
	- Í∑∏Î£π 2. ÌëúÎ≥∏Í≥µÍ∞Ñ $n=8~(R_2=\frac{~8~}{16}),~$ ÌååÎûÄ Í≥µ 5Í∞ú, Îπ®Í∞Ñ Í≥µ 3Í∞ú

$~~~~~~~~~~~\Rightarrow~\text{Gini}=0.5\{1-\left((\frac{~1~}{8})^2 + (\frac{~7~}{8})^2\right)\} + 0.5\{1-\left((\frac{~5~}{8})^2 + (\frac{~3~}{8})^2\right)\}$

```python
0.5*(1 - ((1/8)**2 + (7/8)**2)) + 0.5*(1 - ((5/8)**2 + (3/8)**2))
```

## üî∞ Scikit Learn ÌôúÏö©

> Scikit LearnÏùÑ ÌôúÏö©Ìï¥ Iris Decision TreeÎ•º Íµ¨ÌòÑÌïúÎã§.

- Ïã∏Ïù¥ÌÇ∑Îü∞ÏùÄ ÌòÑÏû¨ ÌååÏù¥Ïç¨ÏóêÏÑú Í∞ÄÏû• Ïú†Î™ÖÌïú Í∏∞Í≥Ñ ÌïôÏäµ Ïò§Ìîà ÏÜåÏä§ ÎùºÏù¥Î∏åÎü¨Î¶¨Ïù¥Îã§.

### ‚èπ Îç∞Ïù¥ÌÑ∞ ÌïôÏäµ --> Î™®Îç∏ ÏÉùÏÑ±

1. Í≤∞Ï†ï Ìä∏Î¶¨Ïùò Î∂ÑÎ•òÍ∏∞(Decision Tree Classifier)Î•º ÏÉùÏÑ±ÌïúÎã§.

2. Î∂ÑÎ•òÍ∏∞Ïóê ÌïôÏäµÌï† Îç∞Ïù¥ÌÑ∞(`features`)ÏôÄ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú Ï†ïÎãµ(`labels`)ÏùÑ ÎÑ£Ïñ¥Ï§ÄÎã§. ‚û° **ÏßÄÎèÑ ÌïôÏäµ (Supervised Learning)**

	- Iris ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ : `iris.data[:, 2:]` --> 3Í∞ú ÌíàÏ¢ÖÏùò 'petal length'ÏôÄ 'petal width' Îç∞Ïù¥ÌÑ∞
	- Iris Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ : `iris.target`

	- Î∂ÑÎ•òÍ∏∞Ïùò Ïù∏Ïàò Îç∞Ïù¥ÌÑ∞Îäî ÌñâÎ†¨ÌòïÌÉú(MatrixLike, `[[]]`)Ïó¨Ïïº ÌïúÎã§.

3. ModelÏù¥ ÎßåÎì§Ïñ¥ ÏßÑÎã§.

```python
iris.data.shape
```

```python
iris.data[:, 2:]
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
iris_tree.fit(iris.data[:, 2:], iris.target)
```

### ‚èπ Î™®Îç∏ --> Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏°

1. ÌïôÏäµÏù¥ ÏôÑÎ£åÎêú Decision Tree Î™®Îç∏(`iris_tree`)Î°ú Îç∞Ïù¥ÌÑ∞Î•º ÏòàÏ∏°(model`.predict`) ÌïúÎã§.

2. Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞(`iris.target`)ÏôÄ ÎπÑÍµêÌï¥ Î≥∏Îã§.

```python
# ÏòàÏ∏° Îç∞Ïù¥ÌÑ∞

y_pred_tree = iris_tree.predict(iris.data[:, 2:])
y_pred_tree
```

```python
# Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞

iris.target
```

### ‚èπ Î™®Îç∏ ÏÑ±Îä•(Accuracy) ÌôïÏù∏

- Scikit LearnÏóêÏÑú Ï†úÍ≥µÌïòÎäî Ï†ïÌôïÎèÑ(Accuracy)Î•º Íµ¨ÌïòÎäî Ìï®Ïàò(`accuracy_score()`)Î•º ÌôúÏö©ÌïúÎã§.

- Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞(`iris.target`)ÏôÄ ÏòàÏ∏° Îç∞Ïù¥ÌÑ∞(`y_pred_tree`)Î•º ÎπÑÍµêÌï¥ Ï†ïÌôïÎèÑ Í∞íÏùÑ Î∞òÌôòÌï¥ Ï§ÄÎã§.

```python
from sklearn.metrics import accuracy_score

# ÏÑ±Îä• ÌôïÏù∏ : Ï†ïÌôïÎèÑ(Accuracy)
accuracy_score(iris.target, y_pred_tree)
```
