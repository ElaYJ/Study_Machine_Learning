## --â–¶ geo + region

```python
geo_city_state = df_geolocation.groupby(by="geolocation_zip_code_prefix", as_index=False)[['geolocation_city', 'geolocation_state']].min()
geo_city_state
```

```python
print(geo_city_state['geolocation_state'].unique().tolist())
len(geo_city_state['geolocation_state'].unique())
```

```python
br_region = dict(
	North=['RR','AP','AM','PA','TO','AC','RO'],
	Northest=['MA','PI','CE','RN','PB','PE','AL','SE','BA'],
	CentralWest=['MT','GO','DF','MS'],
	Southest=['MG','ES','RJ','SP'],
	South=['PR','SC','RS']
)
br_region
```

```python
brazil_region = pd.DataFrame({})

idx = 0
for key, vals in br_region.items():
    for val in vals:
        brazil_region.loc[idx, 'State'] = val
        brazil_region.loc[idx, 'Region'] = key
        idx+=1

brazil_region
```

```python
geo_city_state = geo_city_state.merge(brazil_region, how='left', left_on='geolocation_state', right_on='State')
geo_city_state.drop(columns='State', inplace=True)
geo_city_state
```

```python
geo_lat_lng = df_geolocation.groupby('geolocation_zip_code_prefix', as_index=False)[['geolocation_lat', 'geolocation_lng']].median()
geo_lat_lng
```

```python
df_cm_geolocation = pd.merge(geo_lat_lng, geo_city_state, on='geolocation_zip_code_prefix')
df_cm_geolocation
```

```python
# Customer purchase trend Year-on-Year

query = """
select
month as month_no,
case 
when a.month='01' then 'Jan' 
when a.month='02' then 'Feb'
when a.month='03' then 'Mar'
when a.month='04' then 'Apr'
when a.month='05' then 'May'
when a.month='06' then 'Jun'
when a.month='07' then 'Jul'
when a.month='08' then 'Aug'
when a.month='09' then 'Sep'
when a.month='10' then 'Oct'
when a.month='11' then 'Nov'
when a.month='12' then 'Dec'
else 0 end as month,
sum(case when a.year= '2016' then 1 else 0 end) as Year2016,
sum(case when a.year= '2017' then 1 else 0 end) as Year2017,
sum(case when a.year= '2018' then 1 else 0 end) as Year2018
from
(select 
customer_id,
order_id,
order_delivered_customer_date,
order_status,
strftime('%Y',order_delivered_customer_date) as Year,
strftime('%m',order_delivered_customer_date) as Month
from olist_orders
where order_status= 'delivered' and order_delivered_customer_date is not null
group by customer_id,order_id,order_delivered_customer_date
order by order_delivered_customer_date asc) a
group by month
order by month_no asc
"""

pd.read_sql_query(query, con=conn)
```
